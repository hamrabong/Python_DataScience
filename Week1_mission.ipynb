## 2. Pandas
* 판다스 라이브러리 불러오기
  * import pandas as pd
* DataFrame: column과 row로 이루어진 2차원 데이터 구조
  * dataframe 생성 예시
  * df = pd.DataFrame({"a": [4, 5, 6, 4], "b": [7, 8, 9, 9], "c": [10, 11, 12, 12]}, index = [1, 2, 3, 4])
* series: 1차원 데이터 구조
* dataframe 정렬(sort)
  * df.sort_values(by, axis=0, ascending=True)
  * by: 정렬 기준 
  * axis=0: 행(row) 기준, axis=1: 열(column) 기준
  * ascending=True: 오름차순 정렬, ascending=False: 내림차순 정렬
* 행/열 삭제(drop)
  * df.drop(index="a")
  * index: 삭제할 행
  * df.drop(columns="b")
  * column: 삭제할 열
* 데이터 그룹화 및 연산(groupby)
  * df.groupby(by).agg(["mean", "sum", "count"])
  * by: 정렬 기준
  * agg() 사용해 다중 통계랑 구할 수 있음
* pivot table 생성
  * pd.pivot_table(df, index="a", columns="b", values="c", aggfunc="mean")
  * df: dataframe
  * index: 행 위치에 들어갈 열
  * columns: 열 위치에 들어갈 열
  * values: 데이터로 사용할 열
  * aggfunc: 데이터 집계함수
* 그래프 생성
  * df.plot()

## 3. File Route
* 절대 경로: 고유한 경로, 차상위 디렉토리 포함

  * ../ 현재 위치의 상위 폴더
  * ../../ 두 단계 위의 상위 폴더
* 현재 파일 위치 출력
  * %pwd
* 데이터 폴더의 모든 파일 찾기
  * from glob import glob
  * glob("file_name/*")
* 특정 문자열 포함된 파일 찾기
  * glob("file_name/*str*")
